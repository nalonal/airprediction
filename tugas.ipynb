```python
#import seluruh library yang dibutuhkan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading dataset
df= pd.read_csv('AirQualityUCI.csv', header=None)
df.head()
```

```python
#defining headers(Capitalized)
col=['DATE','TIME','CO_GT','PT08_S1_CO','NMHC_GT','C6H6_GT','PT08_S2_NMHC',
     'NOX_GT','PT08_S3_NOX','NO2_GT','PT08_S4_NO2','PT08_S5_O3','T','RH','AH']
```

```python
#Skiping Row1, Including 'col' i.e. Headers
df1=pd.read_csv('AirQualityUCI.csv',skiprows=1,header=None,names=col)
df1.head(10)
```

```python
#Defining col numbers
r=list(np.arange(len(col)))
```

```python
#Cleaning of Data

#Counting Missing Values from each Column
print('Missing values:\n\n',df.shape[0]-df.count())
clean=(df1.shape[0]-df.count()).sum()
print("\n")
if(clean==0):
    print("No Missing Values")
```

```python
#Defining Month from DATE column
# df1['DATE']=pd.to_datetime(df1.DATE, format='%d-%m-%Y')
# df1['MONTH']= df1['DATE'].dt.month  
# df1.head(10)
df1['DATE']= pd.to_datetime(df1.DATE, format='%d-%m-%Y')
df1['MONTH']= df1['DATE'].dt.month  
df1.head(10)
```

```python
#split hour
df1['HOUR']=df1['TIME'].apply(lambda x: int(x.split(':')[0]))
df1.HOUR.head(10)
```

```python
sns.heatmap(df1.corr(),annot=True, linewidths=.4)
plt.title('Heatmap of co-relation between variables',fontsize=16)
plt.show()
```

```python
#Understanding Linearity between CO & other variables

col1=df1.columns.tolist()[2:]
for i in df1.columns.tolist()[2:]:
    sns.lmplot(x=i,y='CO_GT',data=df1,markers='.')
```

```python
#Linear Regression
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LinearRegression
```

```python
#Let X-input and Y-input features
X=df1[col1].drop('CO_GT',1)     
Y=df1['CO_GT']   
Xsd=sklearn.preprocessing.StandardScaler().fit_transform(X)    #applying stardardisation
X.head()
```

```python
Y.head()
```

```python

X_train, X_test, y_train, y_test=train_test_split(Xsd,Y,test_size=0.4, random_state=3)
```

```python
#Data-Size
print('Training Data Size:',X_train.shape)
print('Test Data Size:',X_test.shape)
```

```python
model= LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
print('Intercept:',model.intercept_)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
y_test
```

```python
print("Co-efficient:")
list(zip(X.columns,model.coef_))
```

```python
#Root Mean Square Error
from sklearn import metrics
rs=np.sqrt(metrics.mean_squared_error(y_test,prediction))
print(rs)
print("\n")
print("Kita dapat menerapkan Regresi Linear untuk memprediksi CO dengan RSME=",rs)
```

```python
print("Logistic Regression")
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()

from sklearn import preprocessing
lab_enc = preprocessing.LabelEncoder()
training_y = lab_enc.fit_transform(y_train)

model1.fit(X_train,training_y)

prediction1=model1.predict(X_test) 
prediction1
```

```python

#Root Mean Square Error
from sklearn import metrics
rs1=np.sqrt(metrics.mean_squared_error(y_test,prediction1))
print(rs1)
print("\n")
print("Dengan menggunakan Logistic Regression kita dapat memprediksi CO dengan RSME=",rs1)
```

```python
print("Decision Tree Regression")
```

```python
from sklearn.tree import DecisionTreeRegressor  
model2=DecisionTreeRegressor()

model2.fit(X_train,y_train)

prediction2=model2.predict(X_test) 
prediction2
```

```python
#Root Mean Square Error
from sklearn import metrics
rs2=np.sqrt(metrics.mean_squared_error(y_test,prediction2))
print(rs2)
print("\n")
print("Dengan menggunakan Decision Tree Regression kita dapat memprediksi CO dengan RSME=",rs2)
```

```python
print("Random Forest Regression")
```

```python
from sklearn.ensemble import RandomForestRegressor          
model3=RandomForestRegressor()
model3.fit(X_train,y_train)
prediction3=model3.predict(X_test) 
prediction3
```

```python
#Root Mean Square Error
from sklearn import metrics
rs3=np.sqrt(metrics.mean_squared_error(y_test,prediction3))
print(rs3)
print("\n")
#Random Forest Regression
print("Dengan menggunakan Random Forest Regression kita dapat memprediksi CO dengan RSME=",rs3)
```

```python
print("Support Vector Machine")
```

```python
from sklearn.svm import SVR          
model4=SVR()

model4.fit(X_train,y_train)

prediction4=model4.predict(X_test) 
prediction4
```

```python

#Root Mean Square Error
from sklearn import metrics
rs4=np.sqrt(metrics.mean_squared_error(y_test,prediction4))
print(rs4)
print("\n")
#Support Vector Machine
print("Dengan menggunakan Support Vector Machine kita dapat memprediksi CO dengan RSME=",rs4)
```
